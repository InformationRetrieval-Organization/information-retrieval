# Preprocess and Build Models
Run this notebook to bootstrap the database, preprocess posts, and build in-memory retrieval models before production startup.

```python
import asyncio
import logging
import os
import sys

repo_root = os.path.abspath(os.path.join(os.getcwd(), ".."))
src_path = os.path.join(repo_root, "src")
if src_path not in sys.path:
    sys.path.insert(0, src_path)

logging.basicConfig(level=logging.INFO, format='%(asctime)s %(levelname)s [%(name)s] %(message)s')
```

```python
from db.session import init_db_schema
from db.helper import init_database
from information_retrieval.globals import init_globals
from preprocessing.preprocessing import preprocess_documents
from information_retrieval.boolean_model import build_boolean_model
from information_retrieval.vector_space_model import (
    build_vector_space_model,
    execute_singualar_value_decomposition,
)
```

```python
async def run_pipeline() -> None:
    await init_db_schema()
    await init_database()

    init_globals()
    await preprocess_documents()
    await build_boolean_model()
    await build_vector_space_model()
    await execute_singualar_value_decomposition()

    print("Preprocessing and model build completed.")

await run_pipeline()
```

## Production
After running this notebook successfully, start the API with `FASTAPI_ENV=production` so startup skips preprocessing and uses the database as the baseline.
